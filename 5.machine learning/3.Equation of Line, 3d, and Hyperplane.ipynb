# 03. Equation of Line, 3D Plane, and Hyperplane

This document summarizes the mathematical definitions and geometric interpretations of lines, planes, and their n-dimensional generalizations (hyperplanes), based on the provided lecture notes.

## 1. Equation of a Line (2D)

In 2D space, a line can be represented in several forms:

### Slope-Intercept Form
$$y = mx + c$$
* **m**: Slope of the line.
* **c**: y-intercept.

### General Form
$$ax + by + c = 0$$
This can be rearranged into the slope-intercept form:
$$by = -ax - c$$
$$y = \left(\frac{-a}{b}\right)x - \left(\frac{c}{b}\right)$$
* Here, slope $m = -\frac{a}{b}$ and intercept $= -\frac{c}{b}$.

### Vector Form
We can represent the variables $(x, y)$ as features $(x_1, x_2)$ and coefficients $(a, b)$ as weights $(w_1, w_2)$.
$$w_1x_1 + w_2x_2 + b = 0$$
$$\mathbf{w}^T\mathbf{x} + b = 0$$
* **$\mathbf{w}$**: Weight vector $\begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$
* **$\mathbf{x}$**: Feature vector $\begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$
* **$b$**: Bias (intercept) term.

---

## 2. Equation of a 3D Plane and Hyperplane (n-Dimensions)

### 3D Plane
Extending the 2D vector form to 3 dimensions:
$$w_1x_1 + w_2x_2 + w_3x_3 + b = 0$$
$$\mathbf{w}^T\mathbf{x} + b = 0$$
where $\mathbf{w} = \begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix}$ and $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}$.

### n-Dimensional Plane (Hyperplane)
Generalizing to $n$ dimensions:
$$w_1x_1 + w_2x_2 + w_3x_3 + \dots + w_nx_n + b = 0$$
$$\mathbf{w}^T\mathbf{x} + b = 0$$
* A hyperplane is a subspace of one dimension less than its ambient space.

---

## 3. Geometric Interpretation (Passing through Origin)

If the hyperplane passes through the origin $(0,0,\dots,0)$, the intercept $b$ becomes 0.
The equation simplifies to:
$$\mathbf{w}^T\mathbf{x} = 0$$

### Orthogonality
The term $\mathbf{w}^T\mathbf{x}$ is the dot product of vectors $\mathbf{w}$ and $\mathbf{x}$:
$$\mathbf{w} \cdot \mathbf{x} = \|\mathbf{w}\| \|\mathbf{x}\| \cos\theta = 0$$

For this to be zero (assuming non-zero vectors), $\cos\theta$ must be 0, which means $\theta = 90^\circ$.
* **Conclusion**: The weight vector $\mathbf{w}$ is always **perpendicular (orthogonal)** to the hyperplane $\Pi$ (or line in 2D).
    * $\mathbf{w} \perp \Pi$
