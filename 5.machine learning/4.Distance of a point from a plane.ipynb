# 04. Distance of a Point from a Plane

This document covers the mathematical concept of calculating the shortest distance between a point and a plane (or Hyperplane in higher dimensions). This concept is fundamental in analytical geometry and has pivotal applications in Machine Learning, such as in Support Vector Machines (SVM) algorithms.

---

## 1. The Core Concept

The distance we are referring to is the **Orthogonal Distance**, which is the shortest distance between a specific point and a plane.

If we have:
* **Point $P$**: Defined by its coordinate vector $\mathbf{x}_0$.
* **Plane (Hyperplane) $\Pi$**: Defined by the linear equation $\mathbf{w}^T\mathbf{x} + b = 0$.

Then the distance $d$ between the point and the plane is given by the following general formula:

$$d = \frac{|\mathbf{w}^T\mathbf{x}_0 + b|}{\|\mathbf{w}\|}$$

Where:
* $\mathbf{w}$: Is the Normal Vector to the plane.
* $b$: Is the Bias term (Intercept).
* $\|\mathbf{w}\|$: Is the Euclidean Norm or the length of the vector $\mathbf{w}$.

---

## 2. Special Cases (from 2D to n-D)

### A. In 2D: Distance of a point from a straight line
In two-dimensional space, a plane is simply a straight line.
* **Line Equation**: $ax + by + c = 0$ (here $\mathbf{w} = [a, b]^T$ and $b_{bias} = c$)
* **Point**: $(x_0, y_0)$
* **Distance Formula**:
    $$d = \frac{|ax_0 + by_0 + c|}{\sqrt{a^2 + b^2}}$$

### B. In 3D: Distance of a point from a plane
In three-dimensional space, we deal with a flat plane.
* **Plane Equation**: $ax + by + cz + d = 0$ (here $\mathbf{w} = [a, b, c]^T$ and $b_{bias} = d$)
* **Point**: $(x_0, y_0, z_0)$
* **Distance Formula**:
    $$d = \frac{|ax_0 + by_0 + cz_0 + d|}{\sqrt{a^2 + b^2 + c^2}}$$

### C. In n-Dimensions: Distance of a point from a Hyperplane
This is the general case used in data science.
* **Hyperplane Equation**: $w_1x_1 + w_2x_2 + \dots + w_nx_n + b = 0$
* **Point**: $\mathbf{x} = [x_1, x_2, \dots, x_n]^T$
* **Distance Formula (Vector notation)**:
    $$d = \frac{|\mathbf{w}^T\mathbf{x} + b|}{\|\mathbf{w}\|}$$
    Where $\|\mathbf{w}\| = \sqrt{w_1^2 + w_2^2 + \dots + w_n^2}$
