### 3 Types of machine learning
Here are the three main types of machine learning, along with their most common algorithms
### üè∑Ô∏è 1. Supervised Learning

This is the most common type of ML. You **teach the algorithm using labeled data**, which means every piece of data already has the correct "answer" or "label" attached. The goal is for the algorithm to learn the mapping function between the inputs and the output.

It's like learning with an answer key.

Supervised learning is mainly split into two types of problems:

#### **A. Classification (Predicting a Category)**
* **Goal:** To predict a discrete class label (e.g., "Spam" or "Not Spam," "Cat" or "Dog," "Good" or "Bad").
* **Common Algorithms:**
    * Logistic Regression
    * K-Nearest Neighbors (KNN)
    * Support Vector Machines (SVM)
    * Naive Bayes
    * Decision Trees
    * Random Forest
    * Gradient Boosting (e.g., XGBoost, LightGBM, CatBoost)

#### **B. Regression (Predicting a Number)**
* **Goal:** To predict a continuous numerical value (e.g., predicting a house price, a person's age, or the temperature tomorrow).
* **Common Algorithms:**
    * Linear Regression
    * Ridge & Lasso Regression (Regularization)
    * Support Vector Regression (SVR)
    * Decision Trees (can also be used for regression)
    * Random Forest (can also be used for regression)

---

### üîç 2. Unsupervised Learning

In this type, you **give the algorithm unlabeled data** (data without any correct answers). The algorithm's job is to find hidden structures, patterns, or relationships within the data on its own.

It's like being given a box of mixed Lego bricks and asked to sort them by color and shape without being told what the colors or shapes are.

Unsupervised learning is mainly split into two types of problems:

#### **A. Clustering (Finding Groups)**
* **Goal:** To automatically group similar data points together based on their features.
* **Common Algorithms:**
    * K-Means
    * DBSCAN
    * Hierarchical Clustering (Agglomerative, Divisive)
    * Mean-Shift

#### **B. Dimensionality Reduction (Simplifying Data)**
* **Goal:** To reduce the number of variables (features) in your dataset while preserving as much important information as possible. This is often used to simplify data before feeding it to a supervised algorithm or for visualization.
* **Common Algorithms:**
    * Principal Component Analysis (PCA)
    * t-SNE (t-distributed Stochastic Neighbor Embedding)
    * Autoencoders (a Deep Learning-based approach)

---

### üéÆ 3. Reinforcement Learning

This type of learning is inspired by behavioral psychology. The algorithm (called an "agent") **learns by interacting with an environment**. It learns through **trial and error**, receiving "rewards" for good actions and "penalties" for bad ones. The agent's goal is to maximize its cumulative reward over time.

It's like training a dog: you give it a treat (reward) when it sits (good action) and a firm "No" (penalty) when it jumps on the couch (bad action).

* **Goal:** To develop an optimal "policy" (a set of rules for making decisions) in a dynamic environment.
* **Common Algorithms & Approaches:**
    * Q-Learning
    * SARSA
    * Deep Q-Networks (DQN) - (This is where Deep Learning and RL combine)
    * Policy Gradient Methods (e.g., REINFORCE)
    * Actor-Critic Methods
